```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando os dados e visualizando os primeiros registros
dados = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv")
dados.head()
```

```python
# Visualizando informações gerais do dataset
dados.info()
```

Pelo comando `dados.info()`, podemos verificar que não há valores nulos, mas para garantir, aplicamos a função `isnull()`.






```python
# Verificando se há valores nulos no dataset

print(dados.isnull().sum())
```

```python
# Definindo a coluna "País" como index para facilitar a manipulação

dados.set_index("País", inplace=True)
dados.head()
```

```python
# Criando uma lista de anos de 1980 a 2013

ano = list(map(str, range(1980, 2014)))
ano
```

```python
# Extraindo os dados de imigração do Brasil para os anos especificados

brasil = dados.loc['Brasil', ano]
print(brasil)
```

```python
# Criando um dicionário com listas de anos e o número de imigrantes brasileiros
brasil_dic = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

# Convertendo o dicionário em um DataFrame
dados_brasil = pd.DataFrame(brasil_dic)
dados_brasil.head()
```

```python
# Plotando o gráfico com o número de imigrantes do Brasil para o Canadá

plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title("Imigrantes do Brasil")
plt.xlabel("Ano")
plt.ylabel("Imigrantes")
plt.show()
```

```python
# Customizando o gráfico para uma melhor visualização

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'],lw=3, color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title("Imigração do Brasil para o Canadá\n 1980 a 2013",fontsize=18, loc='center')
ax.set_xlabel("Ano", fontsize=14)
ax.set_ylabel("Número de imigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
plt.grid(linestyle='--')
plt.savefig('country.png')
plt.show()
```

O gráfico abaixo mostra a imigração do Brasil para o Canadá ao longo dos anos. Vamos observar as tendências de imigração e identificar os períodos com maior fluxo migratório:

- Observa-se um primeiro pico no número de imigrantes por volta de 1992.
- Após isso, a imigração começa a crescer novamente, culminando em um grande pico entre 2005 e 2010.
- Após esse período, há uma nova queda, mas o número de imigrantes permanece superior aos níveis anteriores.





```python
# Agora vamos analisar somente a região da América do Sul

america_do_sul = dados.query('Região == "América do Sul"')
america_do_sul.head()
```

```python
# Visualizando o total de imigrantes dos países da América do Sul entre 1980 e 2013

fig, ax = plt.subplots(figsize=(8,4))

ax.bar(america_do_sul.index, america_do_sul['Total'])
ax.set_title("Total de imigrantes da América do Sul para o Canadá")
ax.set_xlabel("País")
ax.set_ylabel("Total de imigrantes")
plt.show()
```

```python
# Mostrando o gráfico com os eixos invertidos

fig, ax = plt.subplots(figsize=(8,4))

ax.barh(america_do_sul.index, america_do_sul['Total'])
ax.set_title("Total de imigrantes da América do Sul para o Canadá")
ax.set_xlabel("Total de imigrantes")
ax.set_ylabel('')
plt.show()
```

```python
# Ordenando de forma decrescente e ajustando a visualização

america_do_sul_ordenado = america_do_sul.sort_values('Total', ascending=True)

cores = ['royalblue','orange', 'forestgreen','orchid','purple', 'brown','gray','olive','navy', 'teal','tomato']

fig, ax = plt.subplots(figsize=(8,4))

ax.barh(america_do_sul_ordenado.index, america_do_sul_ordenado['Total'], color = cores)
ax.set_title("Total de imigrantes da América do Sul para o Canadá", fontsize = 15)
ax.set_xlabel("Total de imigrantes", fontsize = 11)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 10)
ax.yaxis.set_tick_params(labelsize = 10)
plt.show()
```

- A cima, visualizamos os países da América do Sul e o total de imigrantes ao Canadá entre 1980 e 2013.
- Pode-se observar que a Guiana e a Colômbia lideram os números de imigrantes para o Canadá. Vamos explorar mais detalhadamente cada país para ver se houve picos específicos que justifiquem os altos valores.






```python
# Extraindo os dados de imigração da Guiana e da Colômbia

guiana = dados.loc['Guiana', ano]
colombia = dados.loc['Colômbia', ano]
```

```python
# Criando DataFrames para Guiana e Colômbia

guiana_dic = {'ano': guiana.index.tolist(), 'imigrantes': guiana.values.tolist()}
dados_guiana = pd.DataFrame(guiana_dic)
colombia_dic = {'ano': colombia.index.tolist(), 'imigrantes': colombia.values.tolist()}
dados_colombia = pd.DataFrame(colombia_dic)
```

```python
dados_guiana.head()
```

```python
dados_colombia.head()
```

Abaixo estão os gráficos da imigração da Guiana e da Colômbia para o Canadá, destacando os principais períodos de aumento no fluxo migratório.

```python
# Plotando os gráficos de imigração da Guiana e da Colômbia

fig, axs = plt.subplots(2, 1, figsize=(8, 8))


axs[0].plot(dados_guiana['ano'], dados_guiana['imigrantes'], lw=3, color='b')
axs[0].set_title("Imigração da Guiana para o Canadá\n 1980 a 2013", fontsize=18, loc='center')
axs[0].set_xlabel("Ano", fontsize=14)
axs[0].set_ylabel("Número de imigrantes", fontsize=14)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].xaxis.set_tick_params(labelsize=12)
axs[0].yaxis.set_tick_params(labelsize=12)
axs[0].grid(linestyle='--')

axs[1].plot(dados_colombia['ano'], dados_colombia['imigrantes'], lw=3, color='r')
axs[1].set_title("Imigração da Colômbia para o Canadá\n 1980 a 2013", fontsize=18, loc='center')
axs[1].set_xlabel("Ano", fontsize=14)
axs[1].set_ylabel("Número de imigrantes", fontsize=14)
axs[1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1].xaxis.set_tick_params(labelsize=12)
axs[1].yaxis.set_tick_params(labelsize=12)
axs[1].grid(linestyle='--')

plt.tight_layout()
plt.show()
```

A Guiana apresentou um maior número de imigrantes entre 1980 e 1997, com picos notáveis em 1982, 1987 e 1994. Já a Colômbia, entre 2000 e 2010, apresentou um grande aumento, com pico em 2006, e depois uma leve queda.




No gráfico abaixo, é possível destacar o Brasil como o quarto país com mais imigrantes para o Canadá na América do Sul durante o período de 1980 a 2013.


```python
# Evidenciando o total de imigrantes do Brasil para o Canadá em meio aos outros países da América do Sul
america_do_sul_ordenado = america_do_sul.sort_values('Total', ascending=True)

cores = []

for País in america_do_sul_ordenado.index:
    if País =='Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(8,4))

ax.barh(america_do_sul_ordenado.index, america_do_sul_ordenado['Total'], color = cores)
ax.set_title("Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013", fontsize = 15)
ax.set_xlabel("Total de imigrantes", fontsize = 11)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 10)
ax.yaxis.set_tick_params(labelsize = 10)
plt.show()
```
